```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Data
train_cosim = [0.99950649, 0.88569385, 0.77930392, 0.53073984, 0.49745841, 0.73138615, 0.7506844, 0.29200453, 0.28361821, 0.6096106]
layers = list(range(1, len(train_cosim) + 1))

# Creating the line plot
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.lineplot(x=layers, y=train_cosim, marker='o')

# Adding title and labels
plt.title('ResNet18 on CIFAR-10 Cosine Similarity', fontsize=16)
plt.xlabel('Layer', fontsize=14)
plt.ylabel('Cosine Similarity', fontsize=14)

# Show plot
plt.show()
```

```python
# Calculate the difference between consecutive cosine similarity values
train_cosim_diff = np.diff(train_cosim)

# Adjust the layer numbers for the difference (since diff reduces the number of elements by 1)
layer_diff = layers[:-1]

# Creating subplots for both the original cosine similarity and its difference
plt.figure(figsize=(20, 6))
sns.set(style="whitegrid")

# Plot for the original cosine similarity
plt.subplot(1, 2, 1)
sns.lineplot(x=layers, y=train_cosim, marker='o')
plt.title('ResNet18 on CIFAR-10 Cosine Similarity', fontsize=16)
plt.xlabel('Layer', fontsize=14)
plt.ylabel('Cosine Similarity', fontsize=14)

# Plot for the difference in cosine similarity
plt.subplot(1, 2, 2)
sns.lineplot(x=layer_diff, y=train_cosim_diff, color='r', marker='o')
plt.title('Difference in Cosine Similarity Between Layers', fontsize=16)
plt.xlabel('Layer', fontsize=14)
plt.ylabel('Difference in Cosine Similarity', fontsize=14)

# Show plot
plt.show()
```
